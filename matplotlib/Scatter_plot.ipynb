```python
# Import matplotlib and allow plots to be shown in Jupyter Notebook
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
```

# **1. Scatter plot with plt.plot**
Like line plots, plt.plot can also be used to create scatter plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker= {}'.format(marker))
plt.legend(numpoints = 1)
plt.xlim(0, 1.8);
```

These character codes can be used together with line and color codes to plot points along with a line connecting them:

```python
plt.plot(x, y, '-xk')
```

Additional plt.plot keyword arguments: (Refer plt.plot documentation for full list)

```python
plt.plot(x, y, '-p', color='gray', markersize=15, markerfacecolor='white', markeredgecolor='gray', markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```
